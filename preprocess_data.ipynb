### Handle Missing Values

Dalam proses pembersihan data, beberapa nilai hilang (NaN) ditangani berdasarkan makna kontekstualnya.
```python
def generate_features(df):
    df = df.copy()
    eps = 1e-5  # untuk hindari div/0

    # — 1) Pastikan stime & ltime datetime
    df['stime'] = pd.to_datetime(df['stime'], errors='coerce')
    df['ltime'] = pd.to_datetime(df['ltime'], errors='coerce')

    # — 2) Duration (detik), clip ke >=0
    df['duration'] = (
        df['ltime'] - df['stime']
    ).dt.total_seconds().clip(lower=0)

    # — 3) Rate features (per detik)
    df['src_bps'] = df['sbytes'] / (df['duration'] + eps)
    df['dst_bps'] = df['dbytes'] / (df['duration'] + eps)
    df['src_pps'] = df['spkts'] / (df['duration'] + eps)
    df['dst_pps'] = df['dpkts'] / (df['duration'] + eps)

    # — 4) Ratio features
    df['byte_ratio']       = df['sbytes'] / (df['dbytes'] + eps)
    df['pkt_ratio']        = df['spkts'] / (df['dpkts'] + eps)
    df['load_ratio']       = df['sload']  / (df['dload'] + eps)
    df['jit_ratio']        = df['sjit']   / (df['djit']  + eps)
    df['inter_pkt_ratio']  = df['sintpkt']/ (df['dintpkt']+ eps)
    df['tcp_setup_ratio']  = df['tcprtt'] / (df['synack'] + df['ackdat'] + eps)

    # — 5) Outlier‐robust loss/window ratios
    df['loss_ratio_src']   = df['sloss']  / (df['spkts'] + eps)
    df['loss_ratio_dst']   = df['dloss']  / (df['dpkts'] + eps)
    df['window_ratio']     = df['swin']   / (df['dwin']  + eps)

    # — 6) Aggregates
    df['total_bytes']      = df['sbytes'] + df['dbytes']
    df['total_pkts']       = df['spkts']  + df['dpkts']
    df['total_load']       = df['sload']  + df['dload']
    df['total_jitter']     = df['sjit']   + df['djit']
    df['total_inter_pkt']  = df['sintpkt']+ df['dintpkt']
    df['total_tcp_setup']  = df['tcprtt'] + df['synack'] + df['ackdat']

    # — 7) Interaction features
    df['byte_pkt_interaction_src'] = df['sbytes'] * df['spkts']
    df['byte_pkt_interaction_dst'] = df['dbytes'] * df['dpkts']
    df['load_jit_interaction_src'] = df['sload']  * df['sjit']
    df['load_jit_interaction_dst'] = df['dload']  * df['djit']
    df['pkt_jit_interaction_src']  = df['spkts']  * df['sjit']
    df['pkt_jit_interaction_dst']  = df['dpkts']  * df['djit']

    # — 8) Statistical
    df['mean_pkt_size']    = (df['smeansz'] + df['dmeansz']) / 2
    df['tcp_seq_diff']     = df['stcpb'] - df['dtcpb']

    # — 9) Service & FTP command ratios
    df['ftp_cmd_flow_ratio']  = df['ct_ftp_cmd']      / (df['ct_flw_http_mthd'] + eps)
    df['service_src_ratio']   = df['ct_srv_src']      / (df['ct_state_ttl']     + eps)
    df['service_dst_ratio']   = df['ct_srv_dst']      / (df['ct_state_ttl']     + eps)

    # — 10) Response body rate
    df['res_bdy_rate']        = df['res_bdy_len']     / (df['duration']         + eps)

    # — 11) Time features (dari stime & ltime)
    df['stime_hour']        = df['stime'].dt.hour
    df['stime_weekday']     = df['stime'].dt.weekday
    df['ltime_hour']        = df['ltime'].dt.hour
    df['ltime_weekday']     = df['ltime'].dt.weekday

    return df
```

### Interpretasi

Berdasarkan distribusi label dalam dataset, terlihat bahwa label 0 (normal) mendominasi sebesar 95.16%, sementara label 1 (serangan) hanya mencakup 4.84% dari seluruh data. Ketidakseimbangan (imbalance) ini menunjukkan bahwa dataset sangat didominasi oleh data normal dibanding data serangan. Hal ini berpotensi menjadi masalah serius saat melatih model machine learning, karena model bisa saja cenderung mengklasifikasikan hampir semua data sebagai normal (label 0) untuk mendapatkan akurasi tinggi secara semu, tanpa benar-benar belajar mendeteksi serangan.

Untuk mengatasi permasalahan ini, beberapa strategi perlu diterapkan. Pertama, gunakan metrik evaluasi yang sensitif terhadap ketidakseimbangan kelas seperti precision, recall, F1-score, dan ROC-AUC, bukan hanya mengandalkan akurasi. Kedua, terapkan teknik resampling seperti SMOTE (Synthetic Minority Over-sampling Technique) untuk menyeimbangkan jumlah sampel antar kelas, atau gunakan model yang mendukung class weighting, seperti RandomForestClassifier(class_weight='balanced') atau XGBoost dengan scale_pos_weight. Langkah-langkah ini penting agar model yang dibangun dapat mengenali pola serangan secara lebih efektif, meskipun datanya sangat minoritas.

```python
from sklearn.preprocessing import PowerTransformer, StandardScaler

# ——— A) Subset normal traffic ———
normal_df = train_df[train_df['label'] == 0].copy()
print("Normal only shape:", normal_df.shape)
```

```python
# ——— B) Tentukan kolom numerik (exclude target) ———
num_cols = normal_df.select_dtypes(include=[np.number]).columns.drop(
    ['label', 'attack_cat_enc'], errors='ignore'
).tolist()
```

```python
# ——— C) Winsorizing (median replacement) ———
for col in num_cols:
    q1, q3 = normal_df[col].quantile([0.25, 0.75])
    iqr = q3 - q1
    lower, upper = q1 - 1.5 * iqr, q3 + 1.5 * iqr
    med = normal_df[col].median()
    normal_df[col] = np.where(
        normal_df[col] < lower, med,
        np.where(normal_df[col] > upper, med, normal_df[col])
    )
```

```python
# ——— D) Reduksi Skewness (Yeo–Johnson) ———
# a) Setup transformer & scaler
pt = PowerTransformer(method='yeo-johnson', standardize=False)
scaler = StandardScaler()

# b) Yeo–Johnson untuk semua fitur numerik (kecuali target)
normal_df[num_cols] = pt.fit_transform(normal_df[num_cols])

# c) Standardisasi (zero mean, unit variance)
normal_df[num_cols] = scaler.fit_transform(normal_df[num_cols])
```

```python
# d) Cek hasil transformasi awal
print("After YJ & standardize:")
print(" Skewness (abs > 0.5):")
print(normal_df[num_cols].skew().abs()
      .sort_values(ascending=False)
      .head(), "\n")
print(" Any NaNs? ", normal_df[num_cols].isna().any().any())
print(" Dtypes:", normal_df[num_cols].dtypes.unique(), "\n")
```

```python
# ——— 7) Optional: Log‐Transform on Remaining Heavily‐Skewed Features ———
to_log = ['smeansz', 'dur']
print("Applying log1p to:", to_log)
for col in to_log:
    normal_df[col] = np.log1p(normal_df[col])

# Re‐scale only the log‐transformed cols
normal_df[to_log] = scaler.fit_transform(normal_df[to_log])

# Cek skewness baru
print("New skewness for log‐transformed cols:")
print(normal_df[to_log].skew(), "\n")
```

```python
# ——— 8) Final Check ———
print("Final pipeline check:")
print(" Skewness (abs > 0.5):")
print(normal_df[num_cols].skew().abs()
      .sort_values(ascending=False)
      .head())
print(" Any NaNs? ", normal_df[num_cols].isna().any().any())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# — 1) Definisikan X dan kedua target Y
#   • X: semua fitur kecuali kolom target
#   • y_attack: multiclass target (0…8)
#   • y_label : binary target (0 normal, 1 attack)
X = train_df.drop(columns=['attack_cat_enc', 'label'])
y_attack = train_df['attack_cat_enc']
y_label  = train_df['label']

# — 2) Split data sekaligus untuk kedua target
#    Kita stratify berdasarkan multiclass, agar semua kelas attack terwakili di train/test
X_train, X_test, \
y_train_attack, y_test_attack, \
y_train_label,  y_test_label = train_test_split(
    X, y_attack, y_label,
    test_size=0.2,
    stratify=y_attack,
    random_state=42
)

# — 3) Scaling fitur numeric
numeric_cols = X_train.select_dtypes(include=['int64','float64']).columns
scaler = StandardScaler()
X_train[numeric_cols] = scaler.fit_transform(X_train[numeric_cols])
X_test[numeric_cols]  = scaler.transform(X_test[numeric_cols])

# — 4) Cek bentuk data
print("X_train:", X_train.shape)
print("X_test :", X_test.shape)
print("y_train_attack distribution:\n", y_train_attack.value_counts(normalize=True))
print("y_train_label distribution:\n",  y_train_label.value_counts(normalize=True))
```

```python
from sklearn.metrics import classification_report

# Contoh multiclass
from sklearn.ensemble import RandomForestClassifier
clf_mc = RandomForestClassifier(random_state=42)
clf_mc.fit(X_train, y_train_attack)
print(classification_report(y_test_attack, clf_mc.predict(X_test)))

# Contoh binary
clf_bin = RandomForestClassifier(random_state=42)
clf_bin.fit(X_train, y_train_label)
y_pred_proba = clf_bin.predict_proba(X_test)[:,1]
y_pred = (y_pred_proba > 0.5).astype(int)
print(classification_report(y_test_label, y_pred))
```
